```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data=pd.read_csv("score.csv")
data
```

```python
count=data["Hours"]
count
```

```python
score=data["Scores"]
score
```

```python
model=LinearRegression()
```

```python
x=np.array(count).reshape(-1,1)
```

```python
x
```

```python
y=np.array(score).reshape(-1,1)
y
```

```python
model.fit(x,y)
```

```python
new_hours=int(input("Enter your hours:"))
new_data=np.array(new_hours).reshape(-1,1)
y_score=model.predict(new_data)
print("Your score is:",model.predict(new_data))
```

```python
plt.plot(x,y)
plt.scatter(new_data,y_score,color="red")
plt.xlabel("Hours")
plt.ylabel("Score")
plt.title("Hours VS Score")
plt.show()
```
